The columns with missing values have a very high percentage (89%, 52%, 38%). Filling these column  up might create a strong deviation or variation, hence wee drop the columns 

```python
df = df.drop(["ADDRESSLINE2", "STATE", "POSTALCODE", "TERRITORY"], axis = 1)
df.columns
```

```python
df.head()
```

JOINING THE CONTACT LAST NAME AND FIRST NAME TOGETHER AND DROPPING THE COLUMNS

```python
df["CONTACT_NAME"] = df["CONTACTFIRSTNAME"] + " " + df["CONTACTLASTNAME"]
df.head()
```

```python
df = df.drop(["CONTACTLASTNAME", "CONTACTFIRSTNAME"], axis = 1)
df.head()
```

Moving the contact name closer to customer name

```python
###function for moving columns
def move(column, index):
    
    """
    column = column to be moved
    index = position to be moved to
    
    """

    col = df.pop(column)
    df.insert(index, column, col)
```

```python
move("CONTACT_NAME", 14)
df.head()
```

The Sales column = Priceeach * Quantityordered. But some values do not correspond.
Hence we drop the column and make a new sales column

```python
df = df.drop("SALES",axis = 1)
```

```python
df.head()
```

```python
df["SALES"] = df["QUANTITYORDERED"] * df["PRICEEACH"]
```

```python
df.columns
```

```python
###moving the column to its former position

move("SALES", 3)
```

```python
df.head()
```

MSRP = MANAGEMENTS STANDARD RETAIL PRICE. We  move it closer to the sales as well as the DEAL SIZE

```python
move("MSRP", 1)
move("DEALSIZE", 5)
```

```python
df.head()
```

PRODUCTLINE column indicates the product they bought, hence we move it closer to the MSRP column

```python
move("PRODUCTLINE", 2)
```

```python
df.head()
```

From the data we see that the MSRP and the PRICEEACH are not the same meaning they either sold above or below the MSRP so assuming we want to know how they would have fared if they sold according to the MSRP we have to create an EXPSALES column and a PROFIT/LOSS column to determine if they made a profit/loss on  that sale based on the MSRP and PRICEEACH

```python
df["EXPSALES"] = df["QUANTITYORDERED"] * df["MSRP"]
df["PROFIT/LOSS"] = df["SALES"] - df["EXPSALES"]
```

```python
##MOVING THE JUST CREATED COLUMNS 

move("EXPSALES", 6)
move("PROFIT/LOSS", 8)
df.head()
```

Converting the ORDERDATE to datetime and extracting the month, day and year and then dropping the MONTH_ID

```python
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])
```

```python
df["ORDER_MONTH"] = df["ORDERDATE"].dt.month_name()
df["ORDER_DAY"] = df["ORDERDATE"].dt.day_name()
df["ORDER_DATE"] = df["ORDERDATE"].dt.day
df["ORDER_YEAR"] = df["YEAR_ID"]
```

```python
df.head()
```

```python
df = df.drop(["ORDERDATE", "MONTH_ID", "YEAR_ID"], axis = 1)
```
